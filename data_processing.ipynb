```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_shaker = pd.read_csv("shaker_time.csv")
print(data_shaker.head())
a, b, c = np.polyfit(data_shaker['N'], data_shaker['Time'], 2)
```

```python
plt.scatter(data_shaker['N'], data_shaker['Time'])
plt.xlabel("N, array size")
plt.ylabel("Time, s")
plt.title("Shaker sort")
x = np.linspace(data_shaker['N'][0], data_shaker['N'][len(data_shaker['N']) - 1])
plt.plot(x, a * x ** 2 + b * x + c, color='r')
plt.show()
```

```python
k = (np.polyfit(np.log(data_shaker['N']), np.log(data_shaker['Time']), 1))
k[0]
```
